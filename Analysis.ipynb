```python
import pandas as pd

data = pd.read_csv('out2.out', sep=" ", header=None)
threads = [1, 2, 4, 8, 16, 32]
K = [3, 5, 7, 9, 11, 13, 15]
```

```python
processed = pd.DataFrame(np.zeros((len(K), len(threads))), columns=threads, index=K)
k = 0
for i in range(len(K)):
    for j in range(len(threads)):
        if k < data.shape[0]:
         processed.iloc[i, j] = data.iloc[k][0]
        k += 1
```

```python
processed.plot(figsize=(6,3))
processed.shape
```

```python
fig = plt.figure(figsize=(16, 10))

for i in range(processed.shape[1]):
    #print(processed.iloc[1, i])
    fig.add_subplot(4, 2, i + 1)
    plt.plot(processed.iloc[i])

plt.show()
```

```python
fig = plt.figure(figsize=(16, 10))

for i in range(processed.shape[1]):
    fig.add_subplot(4, 2, i + 1)
    plt.plot(processed.iloc[:, i])

plt.show()
```
